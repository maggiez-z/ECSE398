```python
import os
import os.path
import csv
import matplotlib.pyplot as plt

search_dir = os.getcwd() #get current directory
os.chdir(search_dir)
files = filter(os.path.isfile, os.listdir(search_dir))
files = [os.path.join(search_dir, f) for f in files] #add path to each file
files.sort(key=lambda x: os.path.getmtime(x)) #sort files based on date of modification 
print("We have found", len(files), "files in the current directory")
print(files)

arrayList = []
holderList = []

#Creates empty arrays for each file in the folder
for d in range(0, (len(files) - 2)):
    with open(files[d]) as f:
        reader = csv.reader(f, delimiter=",")
        for i in reader:
            if i[0] != 'TOF' and i[0] != '': 
                holderList.append(float(i[0]))
        print(holderList)
        arrayList.append(sum(holderList)/len(holderList))
        holderList = []
print("This is the averages of each file....")
print(arrayList)

xPoint = []
for c in range(1, (len(arrayList)+1)):
   xPoint.append(c)

print(xPoint)
plt.plot(xPoint,arrayList)
plt.show
```
